```python
import random

while True:
    liczba = random.randint(1, 100)
    proby = 0

    print("Zgadnij liczbę od 1 do 100")

    while True:
        guess = int(input("Twoja liczba: "))
        proby += 1

        if guess < liczba:
            print("Za mało")
        elif guess > liczba:
            print("Za dużo")
        else:
            print("Brawo! Odgadłeś w", proby, "próbach")
            break

    again = input("Czy chcesz zagrać ponownie? (t/n): ")

    if again.lower() != "t":
        print("Koniec gry")
        break
```

```python
def policz_slowa(tekst):
    """Zwraca liczbę słów w tekście."""
    return len(tekst.split())


def najdluzsze_slowo(tekst):
    """Zwraca najdłuższe słowo w tekście."""
    slowa = tekst.split()
    return max(slowa, key=len)


def czestotliwosc_slow(tekst):
    """Zwraca słownik {słowo: liczba_wystąpień}."""
    slowa = tekst.split()
    wynik = {}

    for slowo in slowa:
        if slowo in wynik:
            wynik[slowo] += 1
        else:
            wynik[slowo] = 1

    return wynik


def cenzura(tekst, zakazane_slowa):
    """Zamienia zakazane słowa na ***."""
    slowa = tekst.split()
    wynik = []

    for slowo in slowa:
        if slowo.lower() in zakazane_slowa:
            wynik.append("***")
        else:
            wynik.append(slowo)

    return " ".join(wynik)


tekst = """
Python to popularny język programowania.
Python jest używany w analizie danych i AI.
"""

print("Liczba słów:", policz_slowa(tekst))
print("Najdłuższe słowo:", najdluzsze_slowo(tekst))
print("Częstotliwość słów:", czestotliwosc_slow(tekst))

zakazane = ["python"]
print("Tekst po cenzurze:", cenzura(tekst, zakazane))
```

```python
import csv

suma = 0
liczba = 0
najlepszy_student = None
najwyzsza_ocena = 0

with open("studenci.csv", newline="", encoding="utf-8") as plik:
    reader = csv.DictReader(plik)

    for wiersz in reader:
        ocena = float(wiersz["ocena"])
        suma += ocena
        liczba += 1

        if ocena > najwyzsza_ocena:
            najwyzsza_ocena = ocena
            najlepszy_student = wiersz["imie"] + " " + wiersz["nazwisko"]

srednia = suma / liczba

with open("raport.txt", "w", encoding="utf-8") as raport:
    raport.write(f"Średnia ocen: {srednia}\n")
    raport.write(f"Najlepszy student: {najlepszy_student} ({najwyzsza_ocena})\n")

print("Raport zapisany do pliku raport.txt")
```

```python
zakupy = [
    {"nazwa": "mleko", "cena": 4.50, "ilość": 2},
    {"nazwa": "chleb", "cena": 5.00, "ilość": 1},
    {"nazwa": "masło", "cena": 7.99, "ilość": 1},
    {"nazwa": "jajka", "cena": 12.50, "ilość": 1},
    {"nazwa": "mleko", "cena": 4.50, "ilość": 3},
]

laczny_koszt = sum(produkt["cena"] * produkt["ilość"] for produkt in zakupy)
print("Łączny koszt:", laczny_koszt, "zł")

najdrozszy = max(zakupy, key=lambda x: x["cena"])
print("Najdroższy produkt:", najdrozszy["nazwa"], "-", najdrozszy["cena"], "zł")

unikalne_produkty = set(produkt["nazwa"] for produkt in zakupy)
print("Unikalne produkty:", unikalne_produkty)

drogie_produkty = [produkt["nazwa"] for produkt in zakupy if produkt["cena"] > 5]
print("Produkty droższe niż 5 zł:", drogie_produkty)
```
